```python
import pandas as pd
#régression avec formule
import statsmodels.formula.api as smf
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#chargement des données
Boston_Data = pd.read_csv('Boston_Data.csv',header=0)
```

```python
#dimensions
print(Boston_Data.shape)
#nombre d'observations
n = Boston_Data.shape[0]
#nombre de variables explicatives
p = Boston_Data.shape[1]
print("nombre d'observations",n)
print("nombre de variables explicatives",p,"dont",p-2,"variables explicatives")
```

```python
#on utilise seulement 5 variables explicatives
#X=pd.DataFrame(np.c_[Boston_Data['X3'],Boston_Data['X6'],Boston_Data['X10'],Boston_Data['X11'],Boston_Data['X13']], columns = ['X3','X6','X10','X11','X13'])

#Transformation de notre jeu de données en Data Frame grace à pandas
#X = pd.DataFrame(np.c_[Boston_Data],
                       #columns = Boston_Data.drop(['Unnamed: 0','Y'],axis=1).columns)
X = pd.DataFrame(np.c_[Boston_Data['R1'],Boston_Data['R2'],Boston_Data['R3'],Boston_Data['R4'],Boston_Data['R5'],
                       Boston_Data['X1'],Boston_Data['X2'],Boston_Data['X3'],Boston_Data['X4'],Boston_Data['X5'],
                       Boston_Data['X6'],Boston_Data['X7'],Boston_Data['X8'],Boston_Data['X9'],Boston_Data['X10'],
                       Boston_Data['X11'],Boston_Data['X12'],Boston_Data['X13']],
                       columns = ['R1','R2','R3','R4','R5','X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12','X13'])
Y = Boston_Data['Y']

#base d'apprentissage et base de test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=42)
print("X_train",X_train.shape)
print("X_test",X_test.shape)
print("Y_train",Y_train.shape)
print("Y_test",Y_test.shape)
```

```python
type(X_train)
```

```python
scaler = StandardScaler()
X_train_n = scaler.fit_transform(X_train)
X_test_n = scaler.transform(X_test)
```

```python
type(X_train_n)
```

```python
Boston_Data.drop(['Unnamed: 0','Y'],axis=1).columns
```

```python
X_train_n = pd.DataFrame(np.c_[X_train_n],
                       columns = Boston_Data.drop(['Unnamed: 0','Y'],axis=1).columns)
```

***

##### Ajout de la colonne " Intercept " constituer de 1

```python
X_train_n=sm.tools.tools.add_constant( X_train_n , prepend=True , has_constant='skip')
```

```python
type(X_train_n)
```

```python
X_train_n
```

##### instanciation

```python
reg = sm.OLS(np.array(Y_train), X_train_n.drop(['R3','X7','X3','R1','R2','R5','R4','X2','X10'],axis=1)).fit()
print(reg.summary())
```

# Évaluation du modèle de régression linéaire

```python
X_train_n = X_train_n.drop(['R3','R2','R5','X7','X3','R4','R1','X10','X2'],axis=1)
```

```python
X_train_n
```

```python
X_test_n = pd.DataFrame(np.c_[X_test_n],
                       columns = Boston_Data.drop(['Unnamed: 0','Y'],axis=1).columns)
```

```python
X_test_n = X_test_n.drop(['R3','R2','R5','X7','X3','R4','R1','X10','X2'],axis=1)
```

```python
X_test_n=sm.tools.tools.add_constant( X_test_n , prepend=True , has_constant='skip')
```

```python
X_test_n
```

On utilise deux métriques pour l’évaluation L’erreur quadratique moyenne (l’erreur d’estimation) et le R-square (la qualité du modèle de régression)

```python
# Evaluation du training set
y_train_predict = reg.predict(X_train_n)
error = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
score = r2_score(Y_train, y_train_predict)

print('La performance du modèle sur la base dapprentissage')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(error))
print('le score R2 est {}'.format(score))
print('\n')

# model evaluation du testing set
y_test_predict = reg.predict(X_test_n)
error = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
score = r2_score(Y_test, y_test_predict)
 
print('La performance du modèle sur la base de test')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(error))
print('le score R2 est {}'.format(score))
```
